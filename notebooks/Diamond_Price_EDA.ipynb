### 5. Bivariate Analysis with 'price' as the Target Variable

#### Numerical Features vs. 'price'

```python
import numpy as np

# Filter out rows where x, y, or z are 0 for correlation calculation to avoid distortion
df_filtered = df[(df['x'] != 0) & (df['y'] != 0) & (df['z'] != 0)]

# Calculate the correlation matrix
correlation_matrix = df_filtered.select_dtypes(include=np.number).corr()

# Plotting the heatmap
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt='.2f', linewidths=.5)
plt.title('Correlation Matrix of Numerical Features (excluding 0 dimensions)')
plt.show()
```

**Insight from Correlation Matrix:**
*   **`carat`** shows a very strong positive correlation with `price` (~0.92), as expected. This means as carat weight increases, price generally increases significantly.
*   **`x`, `y`, `z`** (dimensions) also show strong positive correlations with `price` (around 0.88-0.89), which is consistent with their direct relationship to diamond size and thus carat weight.
*   **`depth` and `table`** have very weak correlations with `price` (close to 0), suggesting they don't directly influence price as much as carat or dimensions. This is expected as these are cut-related proportions rather than size.
*   **High multicollinearity** is observed among `carat`, `x`, `y`, and `z`. This is natural as these features are highly interdependent (dimensions determine volume, which affects carat). This will need to be considered in modeling to avoid redundant features or handle multicollinearity.

```python
plt.figure(figsize=(18, 5))

# Scatter plot of Carat vs. Price
plt.subplot(1, 3, 1)
sns.scatterplot(data=df_filtered, x='carat', y='price', alpha=0.5)
plt.title('Carat vs. Price')

# Scatter plot of x (length) vs. Price
plt.subplot(1, 3, 2)
sns.scatterplot(data=df_filtered, x='x', y='price', alpha=0.5)
plt.title('Length (x) vs. Price')

# Scatter plot of y (width) vs. Price
plt.subplot(1, 3, 3)
sns.scatterplot(data=df_filtered, x='y', y='price', alpha=0.5)
plt.title('Width (y) vs. Price')

plt.tight_layout()
plt.show()

plt.figure(figsize=(6, 5))
# Scatter plot of z (depth) vs. Price
sns.scatterplot(data=df_filtered, x='z', y='price', alpha=0.5)
plt.title('Depth (z) vs. Price')
plt.tight_layout()
plt.show()
```

**Insight from Scatter Plots (Numerical vs. Price):**
*   **`carat` vs. `price`:** The scatter plot clearly shows a strong, non-linear positive relationship. As carat increases, the price increases exponentially rather than linearly. There's also a visible clustering of lower-carat, lower-price diamonds, and a wider spread of prices for higher-carat diamonds.
*   **`x`, `y`, `z` vs. `price`:** Similar to carat, the dimensions (`x`, `y`, `z`) also show strong positive and non-linear relationships with price. This is expected given their high correlation with `carat`.
*   **Outliers/Anomalies:** The plots also highlight some potential outliers, such as diamonds with very large dimensions but relatively low prices, or vice versa, which might warrant further investigation.

#### Categorical Features vs. 'price'

```python
plt.figure(figsize=(18, 6))

# Box plot of Cut vs. Price
plt.subplot(1, 3, 1)
sns.boxplot(data=df, x='cut', y='price', order=df['cut'].value_counts().index)
plt.title('Cut vs. Price')

# Box plot of Color vs. Price
plt.subplot(1, 3, 2)
sns.boxplot(data=df, x='color', y='price', order=df['color'].value_counts().index)
plt.title('Color vs. Price')

# Box plot of Clarity vs. Price
plt.subplot(1, 3, 3)
sns.boxplot(data=df, x='clarity', y='price', order=df['clarity'].value_counts().index)
plt.title('Clarity vs. Price')

plt.tight_layout()
plt.show()
```

**Insight from Box Plots (Categorical vs. Price):**
*   **`cut` vs. `price`:** There's a general trend where better cuts ('Ideal', 'Premium', 'Very Good') tend to have higher median prices and a wider range of prices, especially for 'Premium'. However, 'Good' and 'Fair' cuts show lower median prices. This suggests that cut quality is a factor, but not as dominant as carat.
*   **`color` vs. `price`:** As color quality decreases (from D to J), the median price generally tends to decrease. Diamonds with colors 'D', 'E', 'F', 'G' typically command higher prices, while 'H', 'I', 'J' have progressively lower median prices. This confirms color as a significant price determinant.
*   **`clarity` vs. `price`:** A clear trend is visible: as clarity improves (from I1 to IF), the median price generally increases. Diamonds with higher clarity grades (VS1, VVS2, VVS1, IF) show higher median prices and often a broader price range, indicating their higher value. Conversely, 'I1' clarity diamonds have the lowest median prices. This highlights clarity as another crucial factor influencing price.

**Insight:** 'cut': 'Ideal' is the most common, followed by 'Premium' and 'Very Good'. 'Fair' is the least common. 'color': 'G' is the most common color, followed by 'E' and 'F'. 'J' is the least common. 'clarity': 'SI1' and 'VS2' are the most common clarity grades, 'I1' is the least common. These distributions indicate common characteristics of diamonds in the dataset.
